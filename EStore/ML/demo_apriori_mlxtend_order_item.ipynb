```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
# pd.set_option('max_columns', None)
```

```python
data = pd.read_csv('orders_items.csv')
data.head()
```

```python
df = data.iloc[:,3:]
df.head()
```

```python
dataset = df.groupby("order_id")['product_id'].apply(list) 
dataset
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
print("Transactions:")
df
```

```python
print("Check null:")
df.isnull().any()
```

```python
print("Support:")
frequent_itemsets = apriori(df, min_support=0.07, use_colnames=True) # dat nguong dua tren so transaction 
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
confidences = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5) 
print("Confidence:")
confidences
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=5)
print("Lift:")
rules
```
